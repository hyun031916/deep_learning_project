```python
#3D Axes 만들기

import numpy as np
import matplotlib.pyplot as plt

#3D 렌더링에 필요한 라이브러리
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

t = np.linspace(-2*np.pi, 2*np.pi)

X, Y = np.meshgrid(t, t)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

#Figure 객체 만들기
fig = plt.figure(figsize=(6, 6))

#서브플롯 ax 만들기
ax = fig.add_subplot(1, 1, 1, projection="3d")

#플롯하기
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
#곡면 만들기
import numpy as np
import matplotlib.pyplot as plt

#3D 렌더링에 필요한 라이브러리
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

x = y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2)/2) / (2*np.pi)

#Figure 객체 만들기
fig = plt.figure(figsize=(6, 6))

#서브플롯 ax 만들기
ax = fig.add_subplot(1, 1, 1, projection="3d")

#곡면 만들기
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
#3D 히스토그램 만들기

import matplotlib.pyplot as plt
import numpy as np

#3D 렌더링에 필요한 라이브러리
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

#Figure 객체
fig = plt.figure(figsize=(5, 5))

#서브플롯 ax1 만들기
ax = fig.add_subplot(111, projection="3d")

#x, y, z의 위치 결정
xpos = [i for i in range(10)]
ypos = [i for i in range(10)]
zpos = np.zeros(10)

#x, y, z의 증가량 결정
dx = np.ones(10)
dy = np.ones(10)
dz = [i for i in range(10)]

#3D 히스토그램 만들기
ax.bar3d(xpos, ypos, zpos, dx, dy, dz)

plt.show()
```

```python
#3D 산포도 만들기
import numpy as np
import matplotlib.pyplot as plt

#3D 렌더링에 필여힌 라이브러리
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(0)

%matplotlib inline

X = np.random.randn(1000)
Y = np.random.randn(1000)
Z = np.random.randn(1000)

#Figure 객체 만들기
fig = plt.figure(figsize=(6, 6))

#서브플롯 ax 만들기
ax = fig.add_subplot(1, 1, 1, projection="3d")

#X, Y, Z를 1차원으로 변환
x = np.ravel(X)
y = np.ravel(Y)
z = np.ravel(Z)

#3D 산포도 만들기
ax.scatter3D(x, y, z)

plt.show()
```

```python
#3D 그래프에 컬러맵 적용하기

import numpy as np
import matplotlib.pyplot as plt

#컬러맵을 위한 라이브러리
from matplotlib import cm

#3D 렌더링에 필요한 라이브러리
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)

R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

#Figure 객체 만들기
fig = plt.figure(figsize=(6, 6))

#서브플롯 ax 만들기 
ax = fig.add_subplot(1, 1, 1, projection="3d")

#다음을 변경하여 Z 값에 컬러맵 제공하기
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

plt.show()
```

```python
#연습문제
#setosa, versicolor, virginica 세 종(Species)의 붓꽃(iris)에 대해 네 개의 측정값(꽃받침의 길이, 
# 꽃받침의 포, 꽃잎의 길이, 꽃잎의 폭) 사용
import matplotlib.pyplot as plt

import pandas as pd
#url 소스에서 iris 데이터 검색
df_iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
df_iris.columns = ["sepal length", "sepal width", "petal length", "petal width", "class"]
#df_iris는 0행부터 50행 까지 세토사(setosa), 51행부터 100행까지 버시컬러(versicolor), 101행부터 150 행까지 버지니카(virginica)
fig = plt.figure(figsize=(10, 10))

#setosa의 sepal length-sepal width의 관계도 그리기
#라벨은 setosa, 색상은 black 지정
plt.scatter(df_iris.iloc[:50, 0], df_iris.iloc[:50, 1], label="setosa", color="k")

#versicolor의 sepal length - sepal width의 관계도 그리기
#라벨은 versicolor, 색상은 blue 지정
plt.scatter(df_iris.iloc[50:100, 0], df_iris.iloc[50:100, 1], label="versicolor", color="b")

#virginica의 sepal length - sepal width의 관계도 그리기
#라벨은 virginica, 색상은 green 지정
plt.scatter(df_iris.iloc[100:150, 0], df_iris.iloc[100:150,1], label="virginaca", color="g")

#x축의 이름을 sepal length로 하기
plt.xlabel("sepal legnth")

#y축의 이름을 sepal width로 하기
plt.ylabel("sepal width")

#그림 표시하기
plt.legend(loc="best")
plt.grid(True)
plt.show()
```
