```python
#3D Axes 

import numpy as np
import matplotlib.pyplot as plt

#3D   
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

t = np.linspace(-2*np.pi, 2*np.pi)

X, Y = np.meshgrid(t, t)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

#Figure  
fig = plt.figure(figsize=(6, 6))

# ax 
ax = fig.add_subplot(1, 1, 1, projection="3d")

#
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
# 
import numpy as np
import matplotlib.pyplot as plt

#3D   
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

x = y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2)/2) / (2*np.pi)

#Figure  
fig = plt.figure(figsize=(6, 6))

# ax 
ax = fig.add_subplot(1, 1, 1, projection="3d")

# 
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
#3D  

import matplotlib.pyplot as plt
import numpy as np

#3D   
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

#Figure 
fig = plt.figure(figsize=(5, 5))

# ax1 
ax = fig.add_subplot(111, projection="3d")

#x, y, z  
xpos = [i for i in range(10)]
ypos = [i for i in range(10)]
zpos = np.zeros(10)

#x, y, z  
dx = np.ones(10)
dy = np.ones(10)
dz = [i for i in range(10)]

#3D  
ax.bar3d(xpos, ypos, zpos, dx, dy, dz)

plt.show()
```

```python
#3D  
import numpy as np
import matplotlib.pyplot as plt

#3D   
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(0)

%matplotlib inline

X = np.random.randn(1000)
Y = np.random.randn(1000)
Z = np.random.randn(1000)

#Figure  
fig = plt.figure(figsize=(6, 6))

# ax 
ax = fig.add_subplot(1, 1, 1, projection="3d")

#X, Y, Z 1 
x = np.ravel(X)
y = np.ravel(Y)
z = np.ravel(Z)

#3D  
ax.scatter3D(x, y, z)

plt.show()
```

```python
#3D   

import numpy as np
import matplotlib.pyplot as plt

#  
from matplotlib import cm

#3D   
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

t = np.linspace(-2*np.pi, 2*np.pi)
X, Y = np.meshgrid(t, t)

R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

#Figure  
fig = plt.figure(figsize=(6, 6))

# ax  
ax = fig.add_subplot(1, 1, 1, projection="3d")

#  Z   
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

plt.show()
```

```python
#
#setosa, versicolor, virginica  (Species) (iris)    ( , 
#  ,  ,  ) 
import matplotlib.pyplot as plt

import pandas as pd
#url  iris  
df_iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
df_iris.columns = ["sepal length", "sepal width", "petal length", "petal width", "class"]
#df_iris 0 50  (setosa), 51 100 (versicolor), 101 150  (virginica)
fig = plt.figure(figsize=(10, 10))

#setosa sepal length-sepal width  
# setosa,  black 
plt.scatter(df_iris.iloc[:50, 0], df_iris.iloc[:50, 1], label="setosa", color="k")

#versicolor sepal length - sepal width  
# versicolor,  blue 
plt.scatter(df_iris.iloc[50:100, 0], df_iris.iloc[50:100, 1], label="versicolor", color="b")

#virginica sepal length - sepal width  
# virginica,  green 
plt.scatter(df_iris.iloc[100:150, 0], df_iris.iloc[100:150,1], label="virginaca", color="g")

#x  sepal length 
plt.xlabel("sepal legnth")

#y  sepal width 
plt.ylabel("sepal width")

# 
plt.legend(loc="best")
plt.grid(True)
plt.show()
```
