```python
#caution
#사전 실행 코드

import matplotlib.pyplot as plt
import cv2
import numpy as np
import time

%matplotlib inline

def aidemy_imshow(name, img):
    b, g, r = cv2.split(img)
    img = cv2.merge([r, g, b])
    plt.imshow(img)
    plt.show()
    
cv2.imshow = aidemy_imshow
```

```python
#트리밍과 리사이즈
import numpy as np
import cv2

img = cv2.imread("cleansing_data/sample.jpg")

#폭가 높이를 각각 1/3로 리사이즈 하기
my_img = cv2.resize(img, (img.shape[1]//3, img.shape[0] //3))

cv2.imshow("sample", my_img)
```

```python
#회전 및 반전
import numpy as np
import cv2

img = cv2.imread("cleansing_data/sample.jpg")

# warpAffine() 함수 사용에 필요한 행렬 만들기
# 첫 번째 인수는 회전의 중심(여기서는 이미지의 중심 설정)
# 두 번째 인수는 회전 각도(여기서는 180도 설정)
# 세 번째 인수는 배율(여기서는 2배 확대로 설정)
mat = cv2.getRotationMatrix2D(tuple(np.array(img.shape[:2])/2), 180, 2.0)

# 아핀 변환 하기
# 첫 번째 인수는 변환하려는 이미지
# 두 번째 인수는 위에서 생성한 행렬(mat)
# 세 번째 인수는 사이즈
my_img = cv2.warpAffine(img, mat, img.shape[:2])

my_img2 = cv2.flip(img, 0)

cv2.imshow("sample", my_img)
cv2.imshow("sample2", my_img2)
```

```python
#색조 변환 및 색상 반전
import numpy as np
import cv2

img = cv2.imread("cleansing_data/sample.jpg")

for i in range(len(img)):
    for j in range(len(img[i])):
        for k in range(len(img[i][j])):
            img[i][j][k] = 255 - img[i][j][k]
            
cv2.imshow("sample", img)
```
