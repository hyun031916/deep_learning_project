```python
#마커 종류와 색상 설정하기

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

days = np.arange(1, 11)
weight = np.array([10, 14, 18, 20, 18, 16, 17, 18, 20, 17])

#그래프 설정
plt.ylim([0, weight.max()+1])
plt.xlabel("days")
plt.ylabel("weight")

#검은색의 원형 마커를 적용한 꺾은 선 그래프 작성
plt.plot(days, weight, marker="o", markerfacecolor="k")
plt.show()
```

```python
#선 종류와 색상 설정하기

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

days = np.arange(1, 11)
weight = np.array([10, 14, 18, 20, 18, 16, 17, 18, 20, 17])

#그래프 설정
plt.ylim([0, weight.max()+1])
plt.xlabel("days")
plt.ylabel("weight")

#검은색 원형 마커 설정, 파란색 파선을 적용한 꺾은 선 그래프 작성
plt.plot(days, weight, linestyle="--", color="b", marker="o", markerfacecolor="k")
plt.show()
```
