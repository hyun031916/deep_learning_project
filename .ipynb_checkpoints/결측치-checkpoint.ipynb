```python
#리스트와이즈 삭제와 페어와이즈 삭제

import numpy as np
from numpy import nan as NA
import pandas as pd

np.random.seed(0)

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[2, 2] = NA
sample_data_frame.iloc[5:, 3] = NA

#페어와이즈 삭제 방식을 사용하여, 0과 2열을 남기되 NaN 포함하는 행 삭제 후 출력하기
sample_data_frame[[0, 2]].dropna()
```

```python
#결측치 보완

import numpy as np
from numpy import nan as NA
import pandas as pd

np.random.seed(0)

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[6:, 2] = NA

sample_data_frame.fillna(method="ffill")
```

```python
#결측치 보완(평균값 대입법)

import numpy as np
from numpy import nan as NA
import pandas as pd

sample_data_frame = pd.DataFrame(np.random.rand(10, 4))

#일부 데이터 누락시키기
sample_data_frame.iloc[1, 0] = NA
sample_data_frame.iloc[6:, 2] = NA


#fillna로 NaN 부분에 열의 평균값 대입
sample_data_frame.fillna(sample_data_frame.mean())
```
