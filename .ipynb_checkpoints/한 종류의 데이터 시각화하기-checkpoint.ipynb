```python
#그래프에 데이터 표시하기
#matplotlib.pyplot을 plt로 import

import matplotlib.pyplot as plt
import numpy as np

#주피터 노트북에서는 %matplotlib inline을 기술하여 시각화된 출력 결과 볼 수 있음
%matplotlib inline

#np.pi는 원주율(파이)를 나타냄
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

#데이터 x, y를 그래프에 플롯하기
plt.plot(x, y)

plt.show()
```

```python
#그래프의 표시 범위 설정하기

#matplotlib.pyplot을 plt로 import
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#np.pi는 원주율(파이)
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

#y축의 표시 범위를 [0, 1]로 하기
plt.ylim([0, 1])

#데이터 x, y를 그래프에 플롯하기
plt.plot(x, y)
plt.show()
```

```python
#그래프 요소에 이름 설정하기

#matplotlib.pyplot을 plt로 import
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2*np.pi)
y = np.sin(x)

#그래프의 제목 설정
plt.title("y=sin(x)(0 < y < 1)")

#그래프의 x축과 y축의 이름 설정
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#y축의 표시 범위를 [0, 1]로 지정
plt.ylim([0, 1])

#데이터 x, y를 그래프에 플롯하기
plt.plot(x, y)
plt.show()
```

```python
#그래프에 그리드 표시하기

#matplotlib.pyplot plt로 import
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2*np.pi)
y = np.sin(x)

#그래프의 제목 설정
plt.title("y=sin(x)")

#그래프의 x축과 y축 이름 설정
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#그래프에 그리드 표시
plt.grid(True)

#데이터 x, y를 그래프에 플롯
plt.plot(x, y)
plt.show()
```

```python
#그래프 축에 눈금 설정하기

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2*np.pi)
y = np.sin(x)

#그래프 제목 설정
plt.title("y=sin(x)")

#그래프의 x축과 y축 이름 설정
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#그래프의 그리드 표시
plt.grid(True)

#positions와 labels 설정
positions = [0, np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ["0˚", "90˚", "180˚", "270˚", "360˚"]

#그래프의 x 축 눈금 설정
plt.xticks(positions, labels)

#데이터 x, y를 그래프에 플롯하기
plt.plot(x, y)
plt.show()
```
