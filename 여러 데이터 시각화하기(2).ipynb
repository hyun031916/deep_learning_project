```python
# 연습문제
# 서브플롯의 객체를 늘려서 여러 그래프를 동시에 표시할 수 있다.
# 3X2의 그래프 레이아웃을 만들어 상단에는 왼쪽부터 순서대로 y=x1, y=x2, y=x3 그래프를,
# 하단에는 왼쪽부터 순서대로 y=sin(x), y=cos(x), y=tan(x)의 그래프 플롯하기
# 각각의 서브플롯에 적절한 제목 붙이고 그리드 표시하기
# 상단에는 _upper가 붙은 변수를, 하단에는 _lower가 붙은 변수 사용하기
# y=tan(x)만 범위가 다르므로 _tan이 붙은 변수 사용하기

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x_upper = np.linspace(0, 5)
x_lower = np.linspace(0, 2*np.pi)
x_tan = np.linspace(-np.pi/2, np.pi/2)
positions_upper = [i for i in range(5)]
positions_lower = [0, np.pi/2, np.pi, np.pi*3/2, np.pi*2]
positions_tan = [-np.pi/2, 0, np.pi/2]
labels_upper = [i for i in range(5)]
labels_lower = ["0˚","90˚", "180˚", "270˚", "360˚"]
labels_tan = ["-90˚", "0˚", "90˚"]

#Figure 객체 만들기
fig = plt.figure(figsize=(9, 6))

#3X2 레이아웃으로 여러 함수 그래프를 플롯하세요
#서브플롯이 서로 겹치지 않도록 설정
plt.subplots_adjust(wspace=0.5, hspace=0.5)

#상단의 서브플롯 만들기
for i in range(3):
    y_upper = x_upper**(i+1)
    ax = fig.add_subplot(2, 3, i+1)
    
    #서브플롯 ax 그래프에 그리드 표시
    ax.grid(True)
    
    #서브플롯 ax 그래프의 제목 설정
    ax.set_title("$y=x^%i$" % (i+1))
    
    #서브플롯 ax 그래프의 x축과 y축 이름 설정
    ax.set_xlabel("x-axis")
    ax.set_ylabel("y-axis")
    
    #서브플롯 ax 그래프의 x축에 눈금 설정
    ax.set_xticks(positions_upper)
    ax.set_yticks(positions_upper)
    
    #데이터 x, y를 그래프에 플롯하기
    ax.plot(x_upper, y_upper)
    
#하단의 서브플롯 만들기
#사용할 함수와 제목을 리스트에 미리 넣은 뒤 for문으로 처리하기
y_lower_list = [np.sin(x_lower), np.cos(x_lower)]
title_list = ["$y=sin(x)$", "$y=cos(x)$"]
for i in range(2):
    y_lower = y_lower_list[i]
    ax = fig.add_subplot(2, 3, i+4)
    
    #서브플롯 ax 그래프에 그리드 표시
    ax.grid(True)
    
    #서브플롯 ax 그래프의 제목 설정
    ax.set_title(title_list[i])
    
    #서브플롯 ax 그래프의 x축과 y축 이름 설정
    ax.set_xlabel("x-axis")
    ax.set_ylabel("y-axis")
              
    #서브플롯 ax 그래프의 x축에 눈금 설정
    ax.set_xticks(positions_lower)
    ax.set_xticklabels(labels_lower)
    
    #데이터 x, y를 그래프에 플롯
    ax.plot(x_lower, y_lower)
    
#y=tan(x) 그래프의 플롯
ax = fig.add_subplot(2, 3, 6)

#서브플롯 ax 그래프에 그리드 표시
ax.grid(True)
    
#서브플롯 ax 그래프의 제목 설정
ax.set_title("$y=tan(x)$")

#서브플롯 ax 그래프의 x축과 y축 이름 설정
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

#서브플롯 ax 그래프의 x축에 눈금 설정
ax.set_xticks(positions_tan)
ax.set_xticklabels(labels_tan)

#서브플롯 ax 그래프의 y축 범위 설정
ax.set_ylim(-1, 1)

#데이터 x, y를 그래프에 플롯
ax.plot(x_tan, np.tan(x_tan))

plt.show()
```
