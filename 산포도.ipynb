```python
#산포도 만들기
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

np.random.seed(0)
x=np.random.choice(np.arange(100), 100)
y=np.random.choice(np.arange(100), 100)

#산포도 만들기
plt.scatter(x, y)

plt.show()
```

```python
#마커 종류와 색상 설정하기
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

np.random.seed(0)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)

#마커 종류는 사각형, 색상은 검은색으로 설정하여 산포도 만들기
plt.scatter(x, y, marker="s", color="k")

plt.show()
```

```python
#값에 따라 마커 크기 설정하기

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

np.random.seed(0)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)

#z값에 의해 마커 크기가 변하도록 플롯하기
plt.scatter(x, y, s = z)
plt.show()
```

```python
#값에 따라 마커 농도 설정하기

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

np.random.seed(0)

x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)

#z 값에 따라 마커 농도가 파란색 계열로 변하도록 설정
plt.scatter(x, y, c=z, cmap="Blues")
```

```python
#컬러바 표시하기
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

np.random.seed(0)
x = np.random.choice(np.arange(100), 100)
y = np.random.choice(np.arange(100), 100)
z = np.random.choice(np.arange(100), 100)

#z 값에 따라 마커 농도가 파란색 계열로 변하도록 설정
plt.scatter(x, y, c=z, cmap="Blues")

#컬러바 표시하기
plt.colorbar()

plt.show()
```
