```python
#막대그래프 만들기
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1, 2, 3, 4, 5, 6]
y = [12, 41, 32, 36, 21, 17]

#막대그래프 만들기
plt.bar(x, y)

plt.show() 
```

```python
#가로축에 라벨 설정하기
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1, 2, 3, 4, 5, 6]
y = [12, 41, 32, 26, 21, 17]
labels = ["Apple", "Orange", "Banana", "Pineapple", "Kiwifruit", "Strawberry"]

#막대 그래프를 만들고 가로축에 라벨 설정
plt.bar(x, y, tick_label = labels)
plt.show()
```

```python
#누적 막대 그래프 만들기
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = [1, 2, 3, 4, 5, 6]
y1 = [12, 41, 32, 36, 21, 17]
y2 = [43, 1, 6, 17, 17, 9]
labels=["Apple", "Orange", "Banana", "PIneapple", "Kiwifruit", "Strawberry"]

#누적 막대 그래프를 만들고, 가로축에 라벨 설정
plt.bar(x, y1, tick_label = labels)
plt.bar(x, y2, bottom=y1)

#계열에 라벨 지정
plt.legend(("y1", "y2"))
plt.show()
```
