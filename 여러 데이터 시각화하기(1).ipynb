```python
#하나의 그래프에 두 종류의 데이터 표시하기
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)
labels=["90˚", "180˚", "270˚", "360˚"]
positions = [np.pi/2, np.pi, np.pi*3/2, np.pi*2]

#그래프의 제목 설정
plt.title("graphs of trigonometric functions")

#그래프의 x축과 y축 이름 설정
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#그래프에 그리드 표시
plt.grid(True)

#그래프의 x축에 눈금 설정
plt.xticks(positions, labels)

#데이터 x, y1을 그래프에 검은색으로 플롯
plt.plot(x, y1, color="black")

#데이터 x, y2를 그래프에 파란색으로 플롯
plt.plot(x, y2, color="blue")

plt.show()
```

```python
#계열에 라벨 설정하기

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)
labels = ["90˚", "180˚", "270˚", "360˚"]
positions = [np.pi/2, np.pi, np.pi*3/2, np.pi*2]

#그래프의 제목 설정
plt.title("graphs of trigonometric functions")

#그래프의 x축과 y축 이름 설정
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#그래프에 그리드 표시
plt.grid(True)

#그래프의 x축에 눈금 설정
plt.xticks(positions, labels)

#데이터 x, y1을 그래프에 검은색으로 플롯
plt.plot(x, y1, color="k", label="y=sin(x)")

#데이터 x, y2를 그래프에 파란색으로 플롯
plt.plot(x, y2, color="b", label="y=cos(x)")

#계열의 라벨 설정
plt.legend(["y=sin(x)", "y=cos(x)"])

plt.show()
```
